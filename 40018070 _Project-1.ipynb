```python
import os
import openpyxl
import xlsxwriter
import pandas as pd

class Student:
    student_data = pd.DataFrame

    def Student_data(self):
        filename = 'Student_Marks.xlsx'
        isFileExist = os.path.isfile(filename)
        if isFileExist:
            self.student_data = pd.read_excel(filename,
                                              sheet_name='Sheet1')
            self.student_data.columns = ['Student_ID', 'Student_Name',
                                         'Physics_Marks', 'Maths_Marks',
                                         'Chemistry_Marks', 'Biology_Marks']
            self.student_data = self.student_data.iloc[1:, :]
            self.student_data.set_index("Student_ID", inplace=True)
            self.student_data = self.student_data.dropna()
            print(self.student_data)
        else:
            # Create a Pandas dataframe from the data.
            self.student_data = pd.DataFrame(columns=['Student_ID', 'Student_Name',
                                                      'Physics_Marks', 'Maths_Marks',
                                                      'Chemistry_Marks', 'Biology_Marks'])

            self.student_data.set_index("Student_ID", inplace=True)
            # Create a Pandas Excel writer using XlsxWriter as the engine.
            writer = pd.ExcelWriter(filename)

            # Convert the dataframe to an XlsxWriter Excel object.
            self.student_data.to_excel(writer, sheet_name='Sheet1')

            # Close the Pandas Excel writer and output the Excel file.
            writer.save()
        inp = 'Y'
        while inp == 'Y':
            inp = input('Do you want to add more (Y/N)?')
            if inp == 'Y':
                self.inserted_data()

                
    def inserted_data(self):
        student_id = int(input("Enter the Student ID:"))
        student_name = input('Enter Name: ')
        physics = int(input("Enter Physics_Marks: "))
        maths = int(input("Enter Maths_Marks"))
        chemistry = int(input("Enter Chemistry_Marks"))
        biology = int(input("Enter Biology_Marks"))

        insert_data = pd.DataFrame([[student_id, student_name, physics, maths, chemistry, biology]],
                                   columns=['Student_ID', 'Student_Name',
                                            'Physics_Marks', 'Maths_Marks',
                                            'Chemistry_Marks', 'Biology_Marks'])
        insert_data.set_index("Student_ID", inplace=True)

        self.student_data = pd.concat([self.student_data, insert_data])
        print(self.student_data)
        self.student_data.append(insert_data)
        self.student_data.to_excel('Student_Marks.xlsx', startrow=len(self.student_data) + 1, header=False)

    def result(self):
        return self.student_data

    
    def top_scorer(self):
        b_index = self.student_data['Biology_Marks'].astype('float64').idxmax()
        p_index = self.student_data['Physics_Marks'].astype('float64').idxmax()
        m_index = self.student_data['Maths_Marks'].astype('float64').idxmax()
        c_index = self.student_data['Chemistry_Marks'].astype('float64').idxmax()

        print('==================== TOP SCORERS ===================')
        print('Biology : ' + str(self.student_data.loc[b_index].Student_Name) + ' ' + str(
            self.student_data.loc[b_index].Biology_Marks))
        print('Physics : ' + str(self.student_data.loc[p_index].Student_Name) + ' ' + str(
            self.student_data.loc[p_index].Physics_Marks))
        print('Maths : ' + str(self.student_data.loc[m_index].Student_Name) + ' ' + str(
            self.student_data.loc[m_index].Maths_Marks))
        print('Chemistry : ' + str(self.student_data.loc[c_index].Student_Name) + ' ' + str(
            self.student_data.loc[c_index].Chemistry_Marks))

        
    def low_scorer(self):
        b_index = self.student_data['Biology_Marks'].astype('float64').idxmin()
        p_index = self.student_data['Physics_Marks'].astype('float64').idxmin()
        m_index = self.student_data['Maths_Marks'].astype('float64').idxmin()
        c_index = self.student_data['Chemistry_Marks'].astype('float64').idxmin()

        print('==================== LOW SCORERS ===================')
        print('Biology : ' + str(self.student_data.loc[b_index].Student_Name) + ' ' + str(
            self.student_data.loc[b_index].Biology_Marks))
        print('Physics : ' + str(self.student_data.loc[p_index].Student_Name) + ' ' + str(
            self.student_data.loc[p_index].Physics_Marks))
        print('Maths : ' + str(self.student_data.loc[m_index].Student_Name) + ' ' + str(
            self.student_data.loc[m_index].Maths_Marks))
        print('Chemistry : ' + str(self.student_data.loc[c_index].Student_Name) + ' ' + str(
            self.student_data.loc[c_index].Chemistry_Marks))

    def Pass(self):
        count = 0
        for i in self.student_data.index:
            if self.student_data.loc[i, 'Physics_Marks'] < 35:
                continue
            elif self.student_data.loc[i, 'Chemistry_Marks'] < 35:
                continue
            elif self.student_data.loc[i, 'Maths_Marks'] < 35:
                continue
            elif self.student_data.loc[i, 'Biology_Marks'] < 35:
                continue
            else:
                print(self.student_data.loc[i, 'Student_Name'] + ' P ')
                count += 1
        print(count)

        
    def Fail(self):
        count = 0
        for i in self.student_data.index:
            if self.student_data.loc[i, 'Physics_Marks'] < 35:
                print(self.student_data.loc[i, 'Student_Name'] + ' F ')
                count += 1
                
            elif self.student_data.loc[i, 'Chemistry_Marks'] < 35:
                print(self.student_data.loc[i, 'Student_Name'] + ' F ')
                count += 1
                
            elif self.student_data.loc[i, 'Maths_Marks'] < 35:
                print(self.student_data.loc[i, 'Student_Name'] + ' F ')
                count += 1
                
            elif self.student_data.loc[i, 'Biology_Marks'] < 35:
                print(self.student_data.loc[i, 'Student_Name'] + ' F ')
                count += 1
                
            else:
                continue
        print(count)

    def all(self):
        print(self.student_data)
        
    def averages(self):
        
        print(self.student_data.mean())
        
    def suggest(self):
        
        student_name = input('Enter student name: ')
        
        student_dict = self.student_data[self.student_data['Student_Name'] == student_name].to_dict(orient='records')[0]
        
        student_marks_dict = dict((k, student_dict[k]) for k in ['Physics_Marks', 'Chemistry_Marks',
                                                                 'Maths_Marks', 'Biology_Marks'] if k in student_dict)
        
        above50_dict = dict((k, v) for k, v in student_marks_dict.items() if v >= 50.0)
        below50_dict = dict((k, v) for k, v in student_marks_dict.items() if v < 50.0)
        
        print('Subjects for '+ student_name +" ",student_marks_dict.items())
        print('Total marks for '+ student_name +" ",sum(student_marks_dict.values()))
        if bool(above50_dict):
            print(student_name + ' is good in subjects : ', *list(above50_dict.keys()), sep=', ')
        if bool(below50_dict):
            print(student_name + ' can improve in subjects : ', *list(below50_dict.keys()), sep=', ')
                
                
                
student = Student()
student.Student_data()
print(" To get all the student's Results -- press 'a',  For Low_scorer -- press 'l'")
print( "for Top_Scores -- press 'p' , for failues --press 'f', for average scores --press 'c', For student analysis -- press --'s' ")

user_input = input('Select a choice for the results: ')

if user_input == 'a':
    student.all()
elif user_input == 'l':
    student.low_scorer()
elif user_input == 't':
    student.top_scorer()
elif user_input == 'p':
    student.Pass()
elif user_input == 'f':
    student.Fail()
elif user_input == 'c':
    student.averages()
elif user_input == 's':
    student.suggest()
```
